```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = pd.read_csv("../Data/coffee/total-production.csv")

path.head(5)
```

# COMPOSITION OF COFFEE PRODUCTION

* Create a Pie chart with the top 5 producers, as well as the sum for the rest of countries.

* Finally, create two donut charts, one representing Brazil's share of global production in 1990, and one for Brazil's share in 2018.

```python
pie_char = coffee_pivot.loc['2018'].sort_values(ascending=False).to_frame().rename(columns={'2018':'cost'})
pie_char.head(5)
```

```python
pie_char.iloc[:5]
```

```python
pie_data = (pd.concat([pie_char.iloc[:5], 
                       pd.DataFrame(
                           {
                               "cost": pie_char.iloc[5:, 0].sum()
                           }
                           , index=["Rest of world"])]
                     )
           )
pie_data
```

```python
fig, ax = plt.subplots()

ax.pie(
    x=pie_data["cost"][::-1],
     startangle=90,
    labels=pie_data.index[::-1], 
    autopct="%.0f%%",
    explode = (0, 0, 0, 0,0,0.2)
)
ax.set_title("Share of Global Coffee Production 2018")
```

*pie_donut

```python
brazil_vs_others = (coffee_pivot
                    .assign(rest_of_world = coffee_pivot
                            .drop("Brazil", axis=1).sum(axis=1))
                    .loc[:, ["Brazil", "rest_of_world"]]
                    .astype({"Brazil": "float64"})
                   )


brazil_vs_others.round(2).head(1)
```

```python
brazil_vs_others.iloc[0]
```

```python
fig, ax = plt.subplots()

ax.pie(
    brazil_vs_others.iloc[0].sort_values(ascending=False),
    startangle=90,
    labels = ["", ""],
    #autopct='%.0f%%',
    pctdistance=.95,
    colors=["white", "Blue"]
)

hole = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# # Adding Circle in Pie chart
fig.gca().add_artist(hole)

plt.text(
    0, 
    0, 
    f"{round((brazil_vs_others.loc['1990', 'Brazil'] / brazil_vs_others.loc['1990'].sum()*100))}%", 
    ha='center', 
    va='center', 
    fontsize=42
)
  

ax.set_title("Brazil Share of Global Production 1990", fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(
    brazil_vs_others.iloc[-1].sort_values(ascending=False),
    startangle=90,
    labels = ["", ""],
    #autopct='%.0f%%',
    pctdistance=.95,
    colors=["white", "Blue"]
)

hole = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# # Adding Circle in Pie chart
fig.gca().add_artist(hole)

plt.text(
    0, 
    0, 
    f"{round((brazil_vs_others.loc['2018', 'Brazil'] / brazil_vs_others.loc['2018'].sum()*100))}%", 
    ha='center', 
    va='center', 
    fontsize=42
)
  

ax.set_title("Brazil Share of Global Production 2018", fontsize=14)

plt.show()
```

# Coffee Consumption Vs. Prices

Compare coffee prices with coffee consumption in the following countries. 

Is there a relationship between coffee consumption and the price these countries pay?

Plot a dual axis bar chart of price vs consumption totals for these countries.

```python
consumption = pd.read_csv("../Data/coffee/imports.csv")


consumption.head(5)
```

```python
melt_consumption =consumption.melt(id_vars='imports',var_name='year',value_name='price')
melt_consumption
```

```python
first_melt = melt_consumption.pivot_table(index='year',columns='imports',values='price',aggfunc='mean',margins=True).round(2).rename_axis(None, axis=0)
df = first_melt.loc['All'].to_frame().reset_index().drop(147,axis=0).rename(columns={'imports':'country','All':'import'})
df
```

```python
prices = pd.read_csv("../Data/coffee/retail-prices.csv")

prices.head(5)
```

```python
melt_price =prices.melt(id_vars='retail_prices',var_name='year',value_name='price')
melt_price
```

```python
second_melt = melt_price.pivot_table(index='year',columns='retail_prices',values='price',aggfunc='mean',margins=True).rename_axis(None, axis=0)
df2 = second_melt.loc['All'].to_frame().reset_index().drop(14,axis=0).rename(columns={'retail_prices':'country','All':'price'})
df2
```

```python
price_cons =  df.merge(df2,how='inner',left_on='country',right_on='country').reindex(labels=['country','price','import'],axis=1).sort_values(by='import',ascending=False)
price_cons
```

```python
fig, ax = plt.subplots(figsize=(13,8))
width=.35
x=np.arange(13)  # length of index

ax.bar(
    x-width/2, 
    price_cons["import"], 
    width=width,
    label='consumption'
)
ax.set_ylabel("Consumption 60kg package",fontsize=14)
ax.legend(bbox_to_anchor=(0.91,1.0))


ax2 = ax.twinx()
ax2.bar(
    x+width/2, 
    price_cons['price'], 
    width=width, 
    label='Price',
    color='orange'
)
ax2.set_ylabel('Average Price Paid per Bag(USD)',fontsize =14)

ax.set_xticks(x)
ax.set_xticklabels(price_cons['country'],fontsize=9);
ax2.legend()
plt.show()
```
